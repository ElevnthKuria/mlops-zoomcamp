# Q3. Dropping outliers

Next, we need to check the distribution of the duration variable. There are some outliers. Let's remove them and keep only the records where the duration was between 1 and 60 minutes (inclusive).

What fraction of the records left after you dropped the outliers?99

```python
yellow_taxi_feb_df.columns
```

```python

# Join using aliases
yellow_taxi_feb_and_jan = pd.concat([yellow_taxi_jan_df, yellow_taxi_feb_df], axis=0)

print(yellow_taxi_feb_and_jan.head())
 
```

```python
def remove_duration_outliers(df, column, lower_bound=1, upper_bound=60):
  """
  Removes outliers from a DataFrame's 'duration' column and calculates the fraction remaining.

  Args:
      df (pandas.DataFrame): The DataFrame containing the 'datetime' and 'duration' columns.
      column (str): The name of the column containing the duration values (e.g., 'duration').
      lower_bound (int, optional): The minimum duration to keep in minutes (inclusive). Defaults to 1.
      upper_bound (int, optional): The maximum duration to keep in minutes (inclusive). Defaults to 60.

  Returns:
      pandas.DataFrame: The DataFrame with outliers removed.
      float: The fraction of records remaining after outlier removal.
  """

  # Convert timedelta to minutes before comparison
  df[column] = df[column] / pd.Timedelta(minutes=1)

  # Filter based on duration bounds (now in minutes)
  filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

  # Calculate fraction of remaining records
  fraction_remaining = len(filtered_df) / len(df)

  return filtered_df, fraction_remaining

# Example usage (assuming yellow_taxi_jan_df has datetime and duration columns)
result_df, fraction_remaining = remove_duration_outliers(yellow_taxi_jan_and_feb.copy(), 'duration')

print("Resulting DataFrame after removing outliers:\n", result_df)
print(f"Fraction of records remaining: {fraction_remaining:.2f}")

```
